```
##dictionary find player and odi and test run in a list inside dictionary

fr=open('test.txt','r')
line=fr.readline()
lines=fr.readlines()
players={}
for line in lines:
  line=line.strip()
  data=line.split('\t')
  pl=data[0]
  run=float(data[1])
  run1=float(data[2])
  if pl not in players:
    players[pl]=[]
  players[pl].append(run)
  players[pl].append(run1)
print(players)

  
```

```
##find ODI average and write in separate file
def avg(ls):
  sm=0.0
  for i in range(len(ls)):
    sm+=ls[i]
  return sm/(len(ls))
players={}
for line in lines:
  line=line.strip()
  data=line.split('\t')
  pl=data[0]
  run=float(data[1])
  run1=float(data[2])
  if pl not in players:
    players[pl]=[]
  players[pl].append(run)
  ##players[pl].append(run1)
fw=open('sample.txt', 'w')
for player in players:
  average=avg(players[player])
  fw.write(player +"\t")
  fw.write(str(average) +'\n')
fw.close()

 
```
